```python
df = pd.read_csv('elevels.csv', decimal=",")
df2 = pd.read_csv('elevels.csv', decimal=",")
```

```python
for i in range(8):
    ax = df.plot(y=i,title='Energy levels',legend=False,style='o')     
    ax.set_xlabel("tiempo s ")
    ax.set_ylabel("Energía mJ")
    
```

```python
Ya = np.zeros(12)
Ys = np.zeros(12)
for i in range(3,11):
    df1 = df[[str(i)]]
    x = df1.index.values
    y = df1.values
    #yerr =  0.01*np.sqrt(df1)
    ya = np.average(y)
    Ya[i] = ya
    Ys[i] = np.std(y)
    yerr = ya + np.std(y)
        
    ax = plt.errorbar(i,ya,xerr=0, yerr=  np.std(y), fmt='o', markersize=8,capsize=10)
    plt.xlabel('Niveles de energía')
    plt.ylabel('Energía promedio mJ')
```

```python
n_groups = 8
YA=Ya[3:11]
YS=Ys[3:11]

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.45

opacity = 0.4
error_config = {'ecolor': '0.1'}

rects1 = ax.bar(index, YA, bar_width,
                alpha=opacity, color='b',
                yerr=YS,
               error_kw=error_config,)
ax.set_xticklabels(('2', '3', '4', '5', '6', '7', '8', '9', '10'))
ax.set_xlabel('Nivel')
ax.set_ylabel('Energía mJ')
```
